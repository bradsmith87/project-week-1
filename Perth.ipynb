```python
# Dependencies
import pandas as pd
import requests
import json
from pyzomato import Pyzomato 
import json
import numpy as np
import matplotlib.pyplot as plt
from api_key import api_key 
```

```python
#Set Perth City ID
perth_id = 296

#Set Base URL
base_url = f"https://developers.zomato.com/api/v2.1/search?entity_id={perth_id}&entity_type=city"

#Print Results
print(requests.get(base_url, headers={"user-key" : api_key}))
```

```python
#Get Data and Convert into JSON
perth_data = requests.get(base_url, headers={"user-key" : api_key}).json()

#Display as JSON
perth_data
```

```python
#Set Value for Pyzomato
p = Pyzomato(api_key)

#Create an Empty List for Restaurant Names
restaurant_names = []

#Begin a For Loop that iterates 10 times starting from 0th record and increases by increments of 20
for x in range(0,200,20):
    restaurants = p.search(lat=31.9505,lon=115.8605,
                           count=20,
                           sort='rating',
                           order='desc',
                           category=10,
                           start=x)
    print(x)
    restaurant_names.append(restaurants)

#Save Restaurant Names as a JSON file
with open('Top200Perth.json', 'w') as fp:
    json.dump(restaurant_names, fp)
```

```python
# Write list of dictionaries to a json file
with open('Top200Perth', 'w') as fp:
    json.dump(restaurant_names, fp)
```

```python
# Read JSON file of restaurants
with open('Top200Perth.json') as json_file:
    allRestaurants = json.load(json_file)
print(allRestaurants)
```

```python
# Create columns for DataFrame
columns = ["Restaurant Name", "Locality", "City", "Cuisines", "Highlights"]

# Create empty DataFrame with above columns
restaurant_names = pd.DataFrame(columns=columns)
```

```python
#Create Loop to get data from values and  Series that is appended to the DataFrame
for restaurantSet in allRestaurants:
    restaurants = restaurantSet['restaurants']
    for restaurant in restaurants:
        singleRow = [restaurant['restaurant']['name'],
               restaurant['restaurant']['location']['locality'],
               restaurant['restaurant']['location']['city'],
               restaurant['restaurant']['cuisines'],
               restaurant['restaurant']['highlights']]
                     
        newRow = pd.Series(singleRow, index=columns)
        restaurant_names = restaurant_names.append(newRow, ignore_index=True)
                

restaurant_names.head(5)
```

```python
restaurant_names.to_csv(r'Top200Perth.csv', index = False)
```
